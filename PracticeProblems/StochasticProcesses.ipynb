###11
$\text{Cov}(X(2), X(3))$


```python
E_2 = (0*(0.1))+(2*(0.2))+(2*0.3)+(4*0.4)
E_3 = (0*(0.1))+(2*(0.2))+(3*0.3)+(5*0.4)

x2_e = x_2 - E_2
x3_e = x_3 - E_3

print([x2_e[i] for i in range(0,4)])
print([x3_e[i] for i in range(0,4)])

result = x2_e*x3_e
np.average(result)
```
